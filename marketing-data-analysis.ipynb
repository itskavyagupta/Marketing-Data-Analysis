```python
#renaming the 'Income' column to remove extra whitespaces

df.rename(columns = {' Income ':'Income'}, inplace = True)
```

```python
# Converting the 'Income' values to type float.

df['Income']=df['Income'].str.replace(',','')
df['Income']=df['Income'].str.replace('$','')
df['Income']=df['Income'].str.replace(" ","")
df['Income']=df['Income'].astype(float)
df.info()
```

```python
# 3. Check for any missing values in the dataset and handle them using an appropriate method.

df.isnull().sum()
```

```python
# 'Income' column has null values. Filling the null values with the mean of the column.

df['Income'] = df['Income'].fillna(df['Income'].mean())
df.isnull().any()

#On checking for null again, we get a False for the 'Income' column as well now.
```

```python
# 4. Check for any presence of special characters in any variables. If present, clean/replace and change the datatype
# of the variable if required.

import string
alphanum = string.ascii_letters+string.digits+string.whitespace

# Checking all the columns with the data type as 'object' for any special characters.

df.Education.str.strip(alphanum).astype(bool).any()
```

```python
df.Marital_Status.str.strip(alphanum).astype(bool).any()
```

```python
df.Dt_Customer.str.strip(alphanum).astype(bool).any()
```

```python
df.Country.str.strip(alphanum).astype(bool).any()
```

```python
# 'Dt_Customer' has special characters. Converting the column to datetime data type.

df['Dt_Customer']=df['Dt_Customer'].str.replace('-','/')
df['Dt_Customer']=df['Dt_Customer'].str.replace(" ","")
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%m/%d/%y')
df.info()
```

```python
# 5. The Marketing Manager wants to know the 'Age' of the customers. Extract the feature age from the given dataset
# and display the statistical summary of the age?

df['Year_Birth'] = pd.to_datetime(df['Year_Birth'], format='%Y')
df['age'] = (pd.to_datetime('now') - df['Year_Birth']).astype('m8[Y]')
df['age'].describe()
```

```python
# 6. The Marketing manager wants to understand the total amount spent on various products so that we can find what
# percentage of the amount is spent on which product.
# a. Find out the total amount spent by a customer.
# b. Display the Percentage of the amount spent on Wines and other products.

df['total_amount'] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
df.head()

# A new column named 'total_amount' is added which shows the total amount spent by a customer on all products.
```

```python
import matplotlib.pyplot as plt

df[df.columns[9:15]].sum().plot(kind='pie', ylabel=' ', autopct='%1.0f%%', title='Percentages of the total amount spent on Wines and other products', figsize=(11,11), table=True)
plt.show()

# Plotting a pie chart to display the percentage of the total amount spent on Wines and other products.
```

```python
# 7. Being an Analyst understands the total number of purchases made through different channels which can help find
# the percentage these channels contribute.
# a. Find out the total purchases done by a customer through different channels.
# b. Display the percentage of the store and other channelsâ€™ contribution to the total purchases.

df['total_purchase'] = df["NumDealsPurchases"] + df["NumWebPurchases"] + df["NumCatalogPurchases"] + df["NumStorePurchases"]
df.head()

# A new column named 'total_purchases' is added which shows the total number of purchases done by a customer.
```

```python
df[df.columns[15:19]].sum().plot(kind='pie', ylabel=' ', autopct='%1.0f%%', title='Percentage of the store and other channelsâ€™ contribution to the total purchases', figsize=(11,11), table=True)
plt.show()

# Plotting a pie chart to display the percentage of the store and other channelsâ€™ contribution to the total purchases.
```

```python
# 8. The marketing manager wants to understand the performance of different marketing campaigns. Find out which
# marketing campaign is most successful? Use suitable graphs for visualization.

df[df.columns[20:26]].sum().plot(kind='bar', xlabel='Campaign Number', ylabel='Number of times it was accpeted', title='Performance of various campaigns', figsize=(11,11), legend=False)
plt.show()
```

```python
# Thus, as we can see from the chart, the most successful campaign was the last campaign.
```

```python
# 9. The marketing manager wants to understand which products are performing the best and which are performing the
# least in terms of revenue. Being an analyst, analyse the data and plot a suitable graph to display a report on
# revenue generated by different products.

df[df.columns[9:15]].sum().plot(kind='area', title='Revenue generated by different products', figsize=(11,11))
plt.show()
```

```python
# Thus, as we can see from the chart, the product which is performing the best is Wines and the least revenue
# generating product is Fruits.
```

```python
# 10. The team wants to understand if thereâ€™s any pattern between the age of customers and the last campaign
# acceptance rate. Plot a suitable graph to visualize the distribution of the age with respect to customers who accepted
# the last campaign.

df.plot(kind='hist', x='Response', y='age')
plt.show()
```

```python
# 11. The Chief Marketing specialist wants to visually see which Country has the most number of customers who
# accepted the last campaign. What is your approach?

df.groupby(['Country']).size().plot(kind = "barh")
plt.show()
```

```python
# 14. Understand the Education background of the customers who complained in the last 2 years. State the Education
# background of the customers who have registered the most number of complaints.

df.groupby("Education").agg({'Complain':'sum'}).sort_values('Complain', ascending=False)
```
